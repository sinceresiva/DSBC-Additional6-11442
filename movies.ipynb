```python
#Filter out ratings with rating > 4
dfRatings[dfRatings.rating > 4]
```

```python
#Find how many null values, missing values are present. Deal with them. Print out how many rows have been modified.
null_columns=dfRatings.columns[dfRatings.isnull().any()]
null_columns.fillna("No Timestamp")
#print(dfRatings)
rws=dfRatings[dfRatings['timestamp']=="No Timestamp"]
print('No:of rows=',rws.shape[0])
```

```python
#Filter out movies from the movies DataFrame that are of type 'Animation'.
dfMovies[dfMovies.genres.str.contains("Animation")]
```

```python
#Find the average rating of movies.
dfMean=dfRatings.groupby(['movieId'])["rating"].mean().reset_index(name ='Average_Rating')
dfMean
```

```python
#Perform an inner join of movies and tags based on movieId.
pd.merge(dfMovies, dfTags, how='inner', on="movieId")
```

```python
#Print out the 5 movies that belong to the Comedy genre and have rating greater than 4.
```

```python
dfComedy = dfMovies[dfMovies.genres.str.contains("Comedy")]
dfComedy=pd.merge(dfMovies[dfMovies.genres.str.contains("Comedy")],dfRatings[dfRatings.rating>4],how='inner',on=['movieId','movieId'])
dfComedy=dfComedy.groupby(["title"])["rating"].sum()
print(dfComedy.head(5))
```

```python
#Split 'genres' into multiple columns.
dfSplitCols = dfMovies['genres'].str.split('|', 5, expand=True)
print(dfSplitCols)
```

```python
#Extract year from title e.g. (1995).
dfYr = dfMovies["title"].str.extract(r'(\d{4})')
print(dfYr)
```

```python
#Get timestamp for 2015-02-01.
import time
ts=time.mktime(time.strptime('2015-02-01', "%Y-%m-%d"))

#Select rows based on timestamps later than 2015-02-01.
dft=dfRatings[dfRatings['timestamp']>ts]
print(dft)
```
